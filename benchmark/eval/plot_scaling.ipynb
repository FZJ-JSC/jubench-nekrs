```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style("darkgrid")
```

```python
path = "../data/JUWELS-Booster/large_scaling.dat"  # Path to the .dat file generated by JUBE. Requires `--style=csv` option

values = pd.read_csv(path, index_col="nodes")
values.sort_index(inplace=True)
```

```python
values["speedup"] = values["total_solve_time"].iloc[0] / values["total_solve_time"]
values
```

```python
# Solve time plot
plt.xticks(values.index)
plt.xlabel("# of nodes")
plt.ylabel("Speedup")

plt.plot(values.index, values["speedup"],              color="green", marker="o", linestyle="solid",  label="Measured")   # Plot performace
plt.plot(values.index, values.index / values.index[0], color="green",             linestyle="dashed", label="Ideal")      # Plot ideal
plt.legend()

plt.savefig("scaling.svg", format="svg")
```
